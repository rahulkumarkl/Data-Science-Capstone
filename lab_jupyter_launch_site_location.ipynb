```python
# Task 2: Mark the success/failed launches for each site on the map
# Group by site to calculate total launches and success rate
site_summary = spacex_df.groupby(['Launch Site', 'Lat', 'Long'])['class'] \
                        .agg(['mean', 'count']).reset_index()

site_summary.rename(columns={'mean': 'success_rate', 'count': 'total_launches'}, inplace=True)

# Create a base map centered around average coordinates
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
launch_map = folium.Map(location=map_center, zoom_start=3)

# Define a color scale based on success rate
def success_color(rate):
    if rate > 0.8:
        return 'green'
    elif rate > 0.5:
        return 'orange'
    else:
        return 'red'

# Add a circle marker for each site
for _, row in site_summary.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Long']],
        radius=10 + row['success_rate'] * 10,  # bigger circles = higher success
        color=success_color(row['success_rate']),
        fill=True,
        fill_color=success_color(row['success_rate']),
        fill_opacity=0.7,
        popup=folium.Popup(
            f"<b>{row['Launch Site']}</b><br>"
            f"Total Launches: {row['total_launches']}<br>"
            f"Success Rate: {row['success_rate']*100:.1f}%",
            max_width=250
        )
    ).add_to(launch_map)

# Add legend
legend_html = """
<div style="position: fixed;
            bottom: 30px; left: 30px; width: 180px; height: 110px;
            border:2px solid grey; z-index:9999; font-size:14px;
            background-color:white; padding: 10px;">
<b>Success Rate</b><br>
<i style="background:green; color:green;">....</i> > 80%<br>
<i style="background:orange; color:orange;">....</i> 50–80%<br>
<i style="background:red; color:red;">....</i> < 50%
</div>
"""
launch_map.get_root().html.add_child(folium.Element(legend_html))

# Save and show the map
launch_map.save("launch_success_rate_map.html")
launch_map
```

```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Define a function to assign marker color based on launch outcome
def assign_marker_color(outcome):
    if outcome == 1:
        return 'green'   # Success
    else:
        return 'red'     # Failure

# Apply the function to create a new column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
```

```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points 
    on the Earth (specified in decimal degrees)
    Returns distance in kilometers.
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371  # Radius of Earth in km
    return c * r

# Example: Cape Canaveral coordinates
launch_site_coords = (28.5623, -80.5774)

# Add a new column 'Distance_km' in spacex_df
spacex_df['Distance_km'] = spacex_df.apply(
    lambda row: haversine(launch_site_coords[0], launch_site_coords[1],
                          row['Lat'], row['Long']),
    axis=1
)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# Launch site coordinates
launch_site_lat = 28.5623
launch_site_lon = -80.5774

# Closest coastline coordinates
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon,
                                        coastline_lat, coastline_lon)

print(f"Distance to closest coastline: {distance_coastline:.2f} km")
```

```python
from folium.features import DivIcon

# Coordinates for the closest coastline point
coastline_coord = [28.56367, -80.57163]  # [Latitude, Longitude]

# Distance from launch site to coastline in km
distance_km = calculate_distance(launch_site_lat, launch_site_lon,
                                 coastline_coord[0], coastline_coord[1])

# Create a marker with distance shown as HTML
distance_marker = folium.Marker(
    location=coastline_coord,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_km)
    )
)

# Add marker to the map
site_map.add_child(distance_marker)
```
